Brief Description about AI Agent_1:

AI Agent 1 is a Web Search Agent that performs advanced financial research by retrieving, verifying, and analyzing online based information. It is designed to have deep investigative financial research and analysis, fact-checking, data-driven reporting, and trend analysis. In addition, it provides detailed and insightful reports on various financial topics. Information related to the financial market, banking, economic trends, academic journals, publications from industries, and investments serves as a source of agent operation. Experts in the financial sector can use the agent to identify investment opportunities, make informed decisions, and stay informed with market trends. The building block of the agent resides on the Agno framework, which is powered by Groq API with the model "llama-3.3-70b-versatile".
The agent acts as the external knowledge acquisition layer, providing up-to-date financial information.

```python
import textwrap
from agno.agent import Agent
from agno.models.groq import Groq
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.tools.newspaper4k import Newspaper4kTools
from agno.tools import tool as agno_tool_decorator
from google.colab import userdata

# Define a custom DuckDuckGoTools that excludes search_news
class LimitedDuckDuckGoTools:
    """A limited DuckDuckGo toolset that only provides web_search."""
    def __init__(self):
        self._duckduckgo = DuckDuckGoTools()

    @agno_tool_decorator()
    def web_search(self, query: str, max_results: int = 10) -> str:
        """Searches the web using DuckDuckGo.
        Args:
            query: The search query.
            max_results: The maximum number of results to return.
        """
        return self._duckduckgo.web_search(query, max_results)

groq_api_key = userdata.get('GROQ_API_KEY')

# Initialize the research agent with advanced journalistic capabilities
research_agent = Agent(
    model=Groq(id="llama-3.3-70b-versatile", api_key=groq_api_key),

    tools=[LimitedDuckDuckGoTools(), Newspaper4kTools()],
    description=textwrap.dedent("""
        You are an elite research analyst in the financial services domain.
        Your expertise encompasses:

        - Deep investigative financial research and analysis
        - fact-checking and source verification
        - Data-driven reporting and visualization
        - Expert interview synthesis
        - Trend analysis and future predictions
        - Complex topic simplification
        - Ethical practices
        - Balanced perspective presentation
        - Global context integration
    """),
    instructions=textwrap.dedent("""
        Follow strict research -> analysis -> reporting workflow.
        Use ONLY provided tools (which include web search, but NOT news search directly from DuckDuckGo).
    """),
    expected_output=textwrap.dedent("""
        # {Compelling Headline}

        ## Executive Summary
        {Concise overview of key findings and significance}

        ## Background & Context
        {Historical context and importance}
        {Current landscape overview}

        ## Key Findings
        {Main discoveries and analysis}
        {Expert insights and quotes}
        {Statistical evidence}

        ## Impact Analysis
        {Current implications}
        {Stakeholder perspectives}
        {Industry/societal effects}

        ## Future Outlook
        {Emerging trends}
        {Expert predictions}
        {Potential challenges and opportunities}

        ## Expert Insights
        {Notable quotes and analysis from industry leaders}
        {Contrasting viewpoints}

        ## Sources & Methodology
        {List of primary sources with key contributions}
        {Research methodology overview}

        ---
        Research conducted by Financial Agent
        Credit Rating Style Report
        Published: {current_date}
        Last Updated: {current_time}
    """),
    markdown=True,
)
```
